```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
```

```python
def fetch_script_text(url):
    # Attempt to retrieve the webpage content
    try:
        response = requests.get(url)
    except:
        return None
    
    # Parse the HTML content using BeautifulSoup
    html_parser = bs4.BeautifulSoup(response.content, "html.parser")
    
    # Extract the text from the HTML, removing all tags
    extracted_text = html_parser.get_text()
    
    # Return the first 1000 characters of the extracted text
    return extracted_text[:1000]

# Apply the fetch function to the 'link' column to get the first 1000 characters of each script
script_samples = scripts['link'].apply(fetch_script_text)
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
```

```python
# Add a new column 'script_text' to the DataFrame, applying the fetch_script_text function
scripts['script_text'] = scripts['link'].apply(fetch_script_text)

# Save the updated DataFrame to a CSV file
scripts.to_csv('pudding_texts.csv', index=False)
```

```python
import requests
import os
import pandas as pd

# TMDb base URL for images
image_base_url = "https://image.tmdb.org/t/p/w500"  # w500 is the size of the image

# Load the dataset (assuming you already have the 'now_playing_movies.csv')
df = pd.read_csv('now_playing_movies.csv')

# Create a directory to save posters if it doesn't exist
if not os.path.exists('posters'):
    os.makedirs('posters')

# Select the first 10 movies (or a random sample)
movies_sample = df.head(10)

# Download and save posters for each movie
for index, movie in movies_sample.iterrows():
    poster_path = movie.get('poster_path')  # Get the poster path from the movie metadata

    if poster_path:
        # Full URL for the poster
        poster_url = f"{image_base_url}{poster_path}"

        # Send a request to download the poster image
        response = requests.get(poster_url)

        if response.status_code == 200:
            # Save the image as a file
            poster_filename = f"posters/{movie['title'].replace(' ', '_')}_poster.jpg"
            with open(poster_filename, 'wb') as f:
                f.write(response.content)
            print(f"Poster saved: {poster_filename}")
        else:
            print(f"Failed to download poster for {movie['title']}. Status code: {response.status_code}")
    else:
        print(f"No poster available for {movie['title']}.")
```
