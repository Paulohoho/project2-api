```python
def fetch_script_text(url):
    # Attempt to retrieve the webpage content
    try:
        response = requests.get(url)
    except:
        return None
    
    # Parse the HTML content using BeautifulSoup
    html_parser = bs4.BeautifulSoup(response.content, "html.parser")
    
    # Extract the text from the HTML, removing all tags
    extracted_text = html_parser.get_text()
    
    # Return the first 1000 characters of the extracted text
    return extracted_text[:1000]

# Apply the fetch function to the 'link' column to get the first 1000 characters of each script
script_samples = scripts['link'].apply(fetch_script_text)
```

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
import requests
import pandas as pd
import os

# Set up TMDB API
api_key = '635b8a2867493a7bdc4a5ebf3bf736ca'
base_url = 'https://api.themoviedb.org/3'
endpoint = f'{base_url}/movie/now_playing'
params = {
    'api_key': api_key,
    'language': 'en-US',
    'page': 1
}

# Fetch movies in theaters
response = requests.get(endpoint, params=params)
movies = response.json().get('results', [])

# Extract metadata fields
movie_data = []
for movie in movies:
    movie_info = {
        'title': movie.get('title'),
        'release_date': movie.get('release_date'),
        'overview': movie.get('overview'),
        'vote_average': movie.get('vote_average'),
    }
    movie_data.append(movie_info)

# Create a DataFrame
df = pd.DataFrame(movie_data)

# Save the DataFrame to a CSV file
df.to_csv('movies_in_theaters.csv', index=False)

```

```python
#download the movie posters for 10 of these movies and save them to this repository
poster_base_url = 'https://image.tmdb.org/t/p/w500'
os.makedirs('posters', exist_ok=True)

for movie in movie_data[:10]:
    poster_path = movie.get('poster_path')
    if poster_path:
        poster_url = f'{poster_base_url}{poster_path}'
        poster_response = requests.get(poster_url)
        if poster_response.status_code == 200:
            with open(f"posters/{movie['title'].replace('/', '_')}.jpg", 'wb') as f:
                f.write(poster_response.content)

# Display the DataFrame
df
```
